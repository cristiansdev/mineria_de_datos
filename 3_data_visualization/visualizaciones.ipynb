# Actividad 3: Visualización de Datos

Este notebook genera diferentes visualizaciones a partir del dataset limpio de películas IMDB.

## Objetivos
- Cargar el dataset limpio.
- Generar al menos 5 tipos de gráficas.
- Automatizar visualizaciones usando bucles.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo
sns.set(style="whitegrid")
%matplotlib inline
```

```python
df = pd.read_csv('../1_data_cleaning/dataset_limpio.csv')
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['duration'], bins=30, kde=True)
plt.title('Histograma de duración de películas')
plt.xlabel('Duración (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='content_rating', y='imdb_score', data=df)
plt.title('Boxplot de puntuación IMDB por clasificación de contenido')
plt.xticks(rotation=45)
plt.show()
```

```python
top_paises = df['country'].value_counts().nlargest(5)
plt.figure(figsize=(6,6))
top_paises.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribución por país (Top 5)')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='budget', y='imdb_score')
plt.title('Presupuesto vs. Puntuación IMDB')
plt.xlabel('Presupuesto')
plt.ylabel('Puntuación IMDB')
plt.show()
```

```python
variables_numericas = ['duration', 'budget', 'imdb_score']

for var in variables_numericas:
    plt.figure(figsize=(8,4))
    sns.histplot(df[var], bins=30, kde=True)
    plt.title(f'Histograma de {var}')
    plt.xlabel(var)
    plt.ylabel('Frecuencia')
    plt.show()
```

## Conclusión

Se generaron múltiples visualizaciones:
- Histogramas de variables numéricas
- Boxplot comparativo por clasificación
- Pie chart de países
- Scatter plot de presupuesto vs. puntuación
- Visualización automatizada con bucles

Estas gráficas permiten observar tendencias generales y relaciones entre variables del dataset.
